```python
# SOLUTION: Question 3
stmt = """  CREATE VIEW v_distinct_trader_per_exchange AS
            select c.exchange as exchange, count(distinct t.trader_id) as traders
            from stocks c
            join transactions t
            on t.stock_symbol = c.stock_symbol
            group by c.exchange;
            
            Create view v_alpha_first_highest_tran_stock_symbol AS
            SELECT stock_symbol, COUNT(*) AS cnt
            FROM transactions
            GROUP BY stock_symbol
            ORDER BY cnt DESC, stock_symbol
            Limit 1;
            
            Create view v_prices_for_timeseries_plot AS
            SELECT trans_date, price_usd
            FROM transactions t
            join v_alpha_first_highest_tran_stock_symbol v
            on v.stock_symbol = t.stock_symbol
            ORDER BY trans_date;
            
            Create view v_movingaverage_for_timeseries_plot AS
            SELECT t.trans_date, AVG(t.price_usd) OVER
            (ORDER BY t.trans_date ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) AS mov_avg
            FROM transactions t
            join v_alpha_first_highest_tran_stock_symbol v
            on v.stock_symbol = t.stock_symbol
            ORDER BY t.trans_date;
            
            CREATE OR REPLACE VIEW v_stocks_bought_per_exchange_per_month as
            SELECT count(DISTINCT transactions.stock_symbol) AS total_stocks, stocks.exchange, 
            (ltrim(rtrim(to_char(to_timestamp(date_part('month'::text, transactions.trans_date)
            ::text, 'MM'::text), 'Month'::text))) || '-'::text) || date_part('year'::text, 
            transactions.trans_date) AS trans_date, 
            to_number(date_part('month'::text, transactions.trans_date)::text || 
            date_part('year'::text, transactions.trans_date), '99G999D9S') AS orderby_trans_date
            FROM transactions
            JOIN stocks 
            ON stocks.stock_symbol::text = transactions.stock_symbol::text
            WHERE transactions.trans_type::text = 'BUY'::text
            GROUP BY stocks.exchange, ((ltrim(rtrim(to_char(to_timestamp(
            date_part('month'::text, transactions.trans_date)::text, 'MM'::text), 
            'Month'::text))) || '-'::text) || date_part('year'::text, transactions.trans_date)),
            to_number(date_part('month'::text, transactions.trans_date)::text || 
            date_part('year'::text, transactions.trans_date), '99G999D9S')
            ORDER BY to_number(date_part('month'::text, transactions.trans_date)::text || 
            date_part('year'::text, transactions.trans_date), '99G999D9S'), stocks.exchange;
            
            create view v_stocks_bought_per_exchange_per_month_2 AS
            select v3.orderby_trans_date, v3.trans_date, v3.exchange, 
            COALESCE(v4.total_stocks, 0) as total_stocks
            from (
            select distinct v1.orderby_trans_date, v1.trans_date, 
            v2.exchange from v_stocks_bought_per_exchange_per_month v1
            join (select distinct exchange from v_stocks_bought_per_exchange_per_month) v2
            on 1=1
            order by v1.orderby_trans_date) V3
            left outer join (
            select total_stocks, trans_date, exchange, 
            orderby_trans_date from v_stocks_bought_per_exchange_per_month) v4
            on v3.trans_date = v4.trans_date
            and v3.exchange = v4.exchange
            order by v3.orderby_trans_date;

        """

connection.execute(stmt)
```
